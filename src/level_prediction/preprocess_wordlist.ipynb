```python
import re
import pandas as pd
```

```python
with open('words.txt', 'r') as file:
    # Read the contents of the file
    file_contents = file.read()
```

```python
split_contents=file_contents.split('\n')
split_contents
```

```python
dict_of_levels = {}
for i in split_contents:
    level = i[0:2]
    dict_of_levels[level] = i[3:]
dict_of_levels
```

```python
tester = re.sub(r'[0-9]+', '', "do1")
tester
```

```python
print(dict_of_levels.keys())
for level_key,text in dict_of_levels.items():
    clean_text = re.sub(r'\([^)]*\)', '', text)
    clean_text = re.sub(r'[0-9]+', '', clean_text)
    clean_text = clean_text.replace(',', ' ')
    clean_text = clean_text.replace('indefinite article', ' ')
    clean_text = clean_text.replace('belongs to me', ' ')
    clean_text = clean_text.replace('auxiliary', ' ')
    clean_text = clean_text.replace('modal', ' ')
    clean_text = [word for word in clean_text.split() if (not word.endswith('.') and not word.endswith('.,'))]
    dict_of_levels[level_key]=clean_text
dict_of_levels
```

```python
labeled_words = pd.DataFrame(columns=['word', 'level'])

for level_key, words in dict_of_levels.items():
    words_frame = pd.DataFrame(words, columns=['word'])
    level_frame = pd.DataFrame([level_key]*len(words), columns=['level'])
    labeled_words = pd.concat([labeled_words, pd.concat([words_frame, level_frame], axis=1)], axis=0)
labeled_words
```

```python
import csv
labeled_words.to_csv('labeled_words.csv', index=False)
```
